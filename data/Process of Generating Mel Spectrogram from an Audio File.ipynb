Convert an Audio File to Mel Spectrogram

Import libraries

```python
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

Load an audio file

```python
signal, sampleRate = librosa.load("./05/5_12_0.wav")
```

Signal Plot (not needed)

```python
plt.plot(signal)
plt.title("Audio Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
```

Mel Spectrogram

```python
melSpec = librosa.feature.melspectrogram(y=signal, sr=sampleRate,
                                         hop_length=512)  # hop length 512 is better for speech processing
melSpec = librosa.power_to_db(melSpec, ref=np.max)
librosa.display.specshow(melSpec, y_axis="mel", fmax=8000, x_axis="time")
plt.title("Mel Spectrogram")
plt.colorbar(format="%+2.0f dB")
```

Save the Mel Spectrogram as a jpg

```python
# remove default white padding around the image
fig, ax = plt.subplots()
fig.subplots_adjust(0, 0, 1, 1)  # left,bottom,right,top
ax.axis('off')

melSpecImg = librosa.display.specshow(melSpec)
plt.savefig("./output.jpg")
plt.show()
```
